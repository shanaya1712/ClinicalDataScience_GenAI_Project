```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart disease.csv')
```

```python
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)
```

```python
# Assuming the target column is named 'target_binary'
correlations = df.corr()['target_binary'].abs().sort_values(ascending=False)[1:4]
print("Top 3 correlations with target_binary:\n", correlations)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['chol'], bins=20, color='skyblue', edgecolor='black')
plt.title('Cholesterol Distribution')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
